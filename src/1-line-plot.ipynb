### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt

def create_line_plot(x, y):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("x and y must be lists.")
    if len(x) != len(y):
        raise ValueError("x and y must be the same length.")
    
    plt.plot(x, y, marker='o')
    plt.title('Square of x')
    plt.xlabel('x')
    plt.ylabel('y = x^2')
    plt.grid(True)
    plt.show()

# Data for the plot
x = [i for i in range(0, 11)]  # Ensure x is a list
y = [i**2 for i in x]  # Ensure y is a list

create_line_plot(x, y)
```
