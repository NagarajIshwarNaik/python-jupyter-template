### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

def create_subplot_layout():
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    
    x_line = [i for i in range(0, 11)]  # Ensure x is a list
    y_line = [i**2 for i in x_line]  # Ensure y is a list
    axs[0, 0].plot(x_line, y_line, marker='o')
    axs[0, 0].set_title('Line Chart')
    axs[0, 0].set_xlabel('x')
    axs[0, 0].set_ylabel('y = x^2')
    axs[0, 0].grid(True)

    categories_bar = ['A', 'B', 'C', 'D']
    values_bar = [5, 7, 3, 9]
    axs[0, 1].bar(categories_bar, values_bar, color=['blue', 'green', 'red', 'purple'])
    axs[0, 1].set_title('Bar Chart')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('x')
    axs[1, 0].set_ylabel('y = 2x')
    for i in range(len(x_scatter)):
        axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,10), ha='center')

    labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
    sizes_pie = [40, 25, 20, 15]
    axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
    axs[1, 1].set_title('Pie Chart')
    axs[1, 1].axis('equal')

    plt.tight_layout()
    plt.show()

create_subplot_layout()
```
